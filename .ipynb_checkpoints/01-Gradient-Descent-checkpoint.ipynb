```python
import numpy as np
import matplotlib.pyplot as plt

a = np.random.uniform(-5, 5)
b = np.random.uniform(-5, 5)
# Función cuadrática y su derivada
def f(x):
    return x**2 + a*x + b

def df(x):
    return 2*x + a

# Algoritmo de descenso por gradiente
def descenso_gradiente(x0, derivada, alpha, steps):
    xs = [x0]
    x = x0
    for _ in range(steps):
        x = x - alpha * derivada(x)
        xs.append(x)
    return xs

# Ejecutar
trayectoria = descenso_gradiente(x0=5, derivada=df, alpha=0.1, steps=25)

# Preparar valores para graficar
x_vals = np.linspace(-10, 10, 200)  # más ancho para ver bien la parábola
y_vals = f(x_vals)

# Crear figura más grande
plt.figure(figsize=(10, 6))  

plt.plot(x_vals, y_vals, label='f(x) = x²')
plt.plot(trayectoria, [f(x) for x in trayectoria], 'o-', label='Descenso')
plt.title("Descenso por Gradiente sobre f(x) = x²")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```
