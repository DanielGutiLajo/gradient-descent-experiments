import numpy as np
import matplotlib.pyplot as plt


```python
# Gradient Descent Algorithm
def gradient_descent(x0, derivative, learning_rate, steps):
    xs = [x0]
    x = x0
    for _ in range(steps):
        x = x - learning_rate * derivative(x)
        xs.append(x)
    return xs
 
```

```python
# Quadratic Function and Its Derivative
def f(x):
    return x**2

def df(x):
    return 2*x

# Run Gradient Descent
trajectory = gradient_descent(x0=0, derivative=df, learning_rate=0.1, steps=25)

# Prepare values for plotting
x_vals = np.linspace(-6, 6, 100)  # wider range to clearly see the parabola
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='f(x) = x²')
plt.plot(trajectory, [f(x) for x in trajectory], 'o-', label='Descent')
plt.title("Gradient Descent on f(x) = x²")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
a = np.random.uniform(-10, 10)
b = np.random.uniform(-5, 5)

# Quadratic Function and Its Derivative
def f(x):
    return x**2 + a*x + b

def df(x):
    return 2*x + a

# Run Gradient Descent
trajectory = gradient_descent(x0=0, derivative=df, learning_rate=0.1, steps=25)

# Prepare values for plotting
x_vals = np.linspace(-6, 6, 100)  # wider range to clearly see the parabola
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label=f'f(x) = x² + {a:.2f}x + {b:.2f}')
plt.plot(trajectory, [f(x) for x in trajectory], 'o-', label='Descent')
plt.title(f"Gradient Descent on f(x) = x² + {a:.2f}x + {b:.2f}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

# Real and estimated minimum
real_min = -a / 2
estimated_min = trajectory[-1]
# Informative text below the graph
print(f"Real minimum (vertex of the parabola): x = {real_min:.4f}, f(x) = {f(real_min):.4f}")
print(f"Estimated minimum after gradient descent: x = {estimated_min:.4f}, f(x) = {f(estimated_min):.4f}")
```
